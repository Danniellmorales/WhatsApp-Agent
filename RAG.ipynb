Modelo de embeddings **OPENAI**

```python
from langchain_openai import OpenAIEmbeddings

embeddings_model_text = OpenAIEmbeddings(model="text-embedding-3-large")
```

Base de datos Chroma

```python
from langchain_chroma import Chroma

vector_store = Chroma(
    collection_name="my_collection",
    embedding_function=embeddings_model_text,
    persist_directory="./chroma_db"
)
```

Agregar documentos a la Vector Store

```python
from uuid import uuid4

from langchain_core.documents import Document

document_1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
    id=1,
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
    id=2,
)

document_3 = Document(
    page_content="Building an exciting new project with LangChain - come check it out!",
    metadata={"source": "tweet"},
    id=3,
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
    id=4,
)

document_5 = Document(
    page_content="Wow! That was an amazing movie. I can't wait to see it again.",
    metadata={"source": "tweet"},
    id=5,
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
    id=6,
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
    id=7,
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!",
    metadata={"source": "tweet"},
    id=8,
)

document_9 = Document(
    page_content="The stock market is down 500 points today due to fears of a recession.",
    metadata={"source": "news"},
    id=9,
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
    id=10,
)

documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
uuids = [str(uuid4()) for _ in range(len(documents))]

vector_store.add_documents(documents=documents, ids=uuids)
```

Consulta a los **vectores**

```python
results = vector_store.similarity_search(
    "eggs for tomorrow morning",
    k=2,
    filter={"source": "tweet"},
)

for res in results:
  print(f"* {res.page_content} [{res.metadata}]")
```

Actualizar **documentos**

```python
updated_document_1 = Document(
    page_content="I had chocolate chip pancakes and fried eggs for breakfast this tomorrow morning.",
    metadata={"source": "tweet"},
    id=1,
)

updated_document_2 = Document(
    page_content="The weather forecast for tomorrow is sunny and warm, with a high of 82 degrees.",
    metadata={"source": "news"},
    id=2,
)

vector_store.update_document(document_id=uuids[0], document=updated_document_1)
# You can also update multiple documents at once
vector_store.update_documents(
    ids=uuids[:2], documents=[updated_document_1, updated_document_2]
)
```

###Agregar documentos de archivos externos

Base de datos chroma par PDF

```python
from langchain_chroma import Chroma

vector_store_pdf = Chroma(
    collection_name="my_collection_for_pdf",
    embedding_function=embeddings_model_text,
    persist_directory="./chroma_db_pdf"
)
```

Segmentar la informacion

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_core.documents import Document
from uuid import uuid4

# Cargar el contenido del archivo
with open("/content/Documento - Amadeus Español .txt") as f:
    state_of_the_union = f.read()

# Inicializa el splitter con los parámetros que deseas
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=150,
    length_function=len,
    is_separator_regex=False,
)

# Divide el contenido del archivo en fragmentos
text_chunks = text_splitter.split_text(state_of_the_union)

# Crea los documentos con IDs únicos
documents = [
    Document(
        page_content=chunk,
        metadata={"source": "Amadeus Document"},
        id=str(uuid4())
    )
    for chunk in text_chunks
]

# Ejemplo: imprime los documentos generados
for doc in documents:
    print(f"* {doc.id} {doc.page_content} [{doc.metadata}]")

uuids = [str(uuid4()) for _ in range(len(documents))]

vector_store_pdf.add_documents(documents=documents, ids=uuids)
```

Query PDF **Amadeus**

```python
results_pdf = vector_store_pdf.similarity_search(
    "¿¿Cómo puedo acceder a LSM si soy usuario administrador? ",
    k=2,
    filter={'source': 'Amadeus Document'},
)

for res_pdf in results_pdf:
  print(f"* {res_pdf.page_content} [{res_pdf.metadata}]")
```

Actualizar informacion con Metada
